```python
import xgboost
import shap

import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings("ignore")
```

```python
X, y = shap.datasets.boston() #데이터 불러오기
model = xgboost.XGBRegressor() #모델정의
model.fit(X, y) #학습
```

```python
X.head()
```

```python
y[1]
```

```python
explainer = shap.Explainer(model)
shap_values = explainer(X)

shap.plots.waterfall(shap_values[0])
```
